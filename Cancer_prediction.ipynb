```python
X=df.drop('diagnosis', axis = 1)
y = df.diagnosis
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 123)  #42
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

def print_score(clf,X_train,X_test, y_train, y_test, train = True):
    if train:
        pred = clf.predict(X_train)
        print("Train results is : \n ======================")
        print(f"accuracy_score: is {accuracy_score(y_train,pred)}\n")
        print(f"confusion_matrix: \n {confusion_matrix(y_train,pred)}\n")
        
    elif train == False:
        pred = clf.predict(X_test)
        print("test results is : \n ======================")
        print(f"accuracy_score: is {accuracy_score(y_test,pred)}\n")
        print(f"confusion_matrix: \n {confusion_matrix(y_test,pred)}\n")
```

```python
# Applying Decision tree

from sklearn.tree import DecisionTreeClassifier
tree= DecisionTreeClassifier(random_state =42)
tree.fit(X_train, y_train)

print_score(tree, X_train, X_test, y_train, y_test, train = True)
print_score(tree, X_train, X_test, y_train, y_test, train = False)
```

```python
#Applying Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

random_F_classifier = RandomForestClassifier(n_estimators= 200)
random_F_classifier.fit(X_train, y_train)

print_score(random_F_classifier, X_train, X_test, y_train, y_test, train = True)
print_score(random_F_classifier, X_train, X_test, y_train, y_test, train = False)
```

```python
#Applying Random Forest Classifier updated ( Parameter tuning)



random_F_classifier_u = RandomForestClassifier(criterion='entropy',max_features= 10, n_estimators= 200)
random_F_classifier_u.fit(X_train, y_train)

print_score(random_F_classifier_u, X_train, X_test, y_train, y_test, train = True)
print_score(random_F_classifier_u, X_train, X_test, y_train, y_test, train = False)
```

```python
#Applying Logistics Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state = 123, penalty='l2', max_iter=200)
model.fit(X_train, y_train)


print_score(model, X_train, X_test, y_train, y_test, train = True)
print_score(model, X_train, X_test, y_train, y_test, train = False)
```

```python
 # Applying Bagging Classifier
from sklearn.ensemble import BaggingClassifier


bgc = BaggingClassifier(DecisionTreeClassifier(), max_samples= 0.5, max_features= 1.0, n_estimators= 20, random_state=123)
bgc.fit(X_train,y_train)
```

```python
print_score(bgc, X_train, X_test, y_train, y_test, train = True)
print_score(bgc, X_train, X_test, y_train, y_test, train = False)
```

```python
#apply gradient boost

from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
gboost_model = GradientBoostingClassifier()
```

```python
gboost_model.fit(X_train,y_train)

print_score(gboost_model, X_train, X_test, y_train, y_test, train = True)
print_score(gboost_model, X_train, X_test, y_train, y_test, train = False)
```

```python
#apply adaptive boost
ada_model = AdaBoostClassifier()
```

```python
ada_model.fit(X_train,y_train)

## predicting Accuracy of the model

print_score(ada_model, X_train, X_test, y_train, y_test, train = True)
print_score(ada_model, X_train, X_test, y_train, y_test, train = False)
```
